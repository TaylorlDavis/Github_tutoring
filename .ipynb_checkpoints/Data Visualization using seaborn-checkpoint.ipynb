```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

# barplot and countplot

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y = 'total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data = tips)
```

# boxplot and violinplot

```python
sns.boxplot(x='day',y="total_bill", data=tips,palette ='Set1')
```

```python
sns.boxplot(data=tips,palette='Set1',orient='h')
```

```python
tips['percent'] = (tips['tip']/tips['total_bill'])*100
```

```python
tips.head(1)
```

```python
sns.boxplot(x='day', y="percent", hue="sex", data=tips, palette="coolwarm")
```

```python
sns.boxplot(x='day', y='tip', hue='sex', data=tips, palette = 'coolwarm')
```

# violinplot

```python
sns.violinplot(x="day", y = 'total_bill', data=tips, palette='rainbow')
```

```python
sns.violinplot(x='day', y='percent', data=tips, hue='sex', palette='Set3')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex',split=True, palette='Set1')
```

# stripplot and swarmplot

```python
sns.stripplot(x='day',y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill',split=True, data=tips, palette='Set1', hue ='sex')
```

```python
sns.violinplot(x='day',y='tip', data=tips,palette = 'rainbow')
sns.swarmplot(x='tip',y='tip', data=tips,color='black', size=3)
```
