```python
import pandas as pd
import os
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
```

```python
names_dir = '/Users/taylo/desktop/baby_names/'
names_data_list =[]
names_columns = ['state', 'gender', 'year', 'name','babies']
for file in os.listdir(names_dir):
    if file.endswith('.TXT'):
        names_data = pd.read_csv(names_dir + file,header=None, names=names_columns)
        names_data_list.append(names_data)
    else:
        continue
names_data = pd.concat(names_data_list)
```

```python
data = names_data[names_data['year'] >= 1950]
```

```python
names_data.head()
```

```python
data.head()
```

```python
data.groupby('gender')['babies'].sum() / data['babies'].sum()
```

```python
data.gender = data['gender']
```

```python
total_babies = data.babies.sum()
female_percent = data[data.gender=='F']['babies'].sum()/total_babies
male_percent = data[data.gender =='M']['babies'].sum()/total_babies
```

```python
male_percent
```

```python
female_percent
```

```python
most_popular_names = (
    data
    .groupby(['name', 'gender'])
    .babies
    .sum()
    .reset_index()
    .sort_values('babies', ascending=False)
)
```

```python
most_popular_names[:5]
```

```python
unique_names_per_gender = (
data
.groupby(['year','gender'])
.name
.nunique()
.unstack()
)
```

```python
unique_names_per_gender.plot(figsize=(10,8), lw=3, color=['crimson','blue'])
sns.despine()
plt.xlabel('')
plt.legend([],frameon = False)
plt.ylabel('Unique Babies Names\n in The Data Set',
          position=(-1,0.8),
           rotation=0,
          ha='left',
          labelpad=145,
          fontsize=14,
          color='gray'
          );
```
