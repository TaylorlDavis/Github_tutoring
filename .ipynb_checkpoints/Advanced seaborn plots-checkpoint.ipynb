```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris_df = sns.load_dataset('iris')
```

```python
iris_df.head(1)
```

```python
sns.scatterplot(x='sepal_width', y='sepal_length',
               data=iris_df,hue='species')
```

```python
palette_dict = {x:'gray' for x in iris_df['species'].unique()}
```

```python
palette_dict
```

```python
palette_dict['virginica']='red'
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='sepal_width',y='sepal_length',
               data=iris_df,hue='species',palette=palette_dict,
               size='petal_length')
```

```python
sns.scatterplot(x='sepal_width', y='sepal_length',
               data=iris_df, hue='species',palette='cubehelix')
```

```python
g=sns.PairGrid(iris_df)
g.map_offdiag(sns.histplot);
g.map_diag(sns.kdeplot)
plt.show()
```

```python
green=sns.light_palette('green')[4]
```

```python
red=sns.light_palette('red')[-1]
```

```python
b=sns.light_palette('blue')[4]
```

```python
g=sns.PairGrid(iris_df)
g.map_lower(sns.scatterplot,color=green);
g.map_diag(sns.boxplot,color=b);
g.map_upper(sns.kdeplot,color=red)
```

```python
g=sns.PairGrid(iris_df,corner=True)
g.map_offdiag(sns.scatterplot);
g.map_diag(sns.boxplot);
```

```python
g=sns.PairGrid(iris_df,hue='species',diag_sharey=False,
              vars=['petal_length','petal_width'])
g.map_diag(sns.kdeplot);
g.map_offdiag(sns.scatterplot)
```

```python
g=sns.PairGrid(iris_df,
               diag_sharey=False,
              x_vars=['sepal_length','sepal_width','petal_width'],
              y_vars=['petal_length'])
g.map_offdiag(sns.scatterplot)
```

```python
g= sns.PairGrid(iris_df,hue='species', diag_sharey=False)
g.map_diag(sns.kdeplot,fill=True,lw=3,ls='--')
g.map_offdiag(sns.scatterplot)
```
