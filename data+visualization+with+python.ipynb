```python
from matplotlib import pyplot as plt
%matplotlib inline
```

$y = x^2$

```python
x = np.linspace(-4,4,200)
y = x**2
```

```python
#linewidth = lw
plt.plot(x,y, color='r', linestyle='--', lw=3)
plt.xlabel('x')
plt.ylabel('y = x**2')
plt.title('Y as a function of x')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1]) #[a,b,c,d] a= distance from left, b= distance from right,c=height d = width
ax2 = fig.add_axes([0.2,0.2,0.4,0.5])
ax1.plot(x,y, 'r')
ax2.plot(y,x, 'g')
```
